```python
import flask
from flask import request, jsonify
import requests
import re
import json

def get_data(key, value, current):
    results = []
    pattern_dict = {
        'C'      : r'\bC\b',
        'C++'    : r'C\+\+',
        'Java'   : r'\bJava\b',
        'C#'     : r'C\#',
        'Python' : r'Python',
        'Scala' : r'Scala',
        'Oracle' : r'Oracle',
        'SQL Server': r'SQL Server',
        'MySQL Server' : r'MySQL Server',
        'PostgreSQL': r'PostgreSQL',
        'MongoDB'    : r'MongoDB',
        'JavaScript' : r'JavaScript',
        'Los Angeles' : r'Los Angeles',
        'New York':'New York',
        'San Francisco':'San Francisco',
        'Washington DC':'Washington DC',
        'Seattle':'Seattle',
        'Austin':'Austin',
        'Detroit':'Detroit'
    }
    
    for rec in current:
        if re.search(pattern_dict[value], str(rec.get(key, ""))) is not None:
            results.append(rec)
    return results


app = flask.Flask(__name__)

# Load job data
with open("jobs.json", encoding="utf-8") as f:
    data = json.load(f)


@app.route('/', methods=['GET'])
def home():
    return "<h1>Welcome to Flask JOB Search API</h1>"


@app.route('/data/all', methods=['GET'])
def api_all():
    return jsonify(data)


@app.route('/data', methods=['GET'])
def api_id():
    res = None
    
    for req in request.args:
        key = req
        value = request.args[req]
        
        if res is None:
            res = get_data(key, value, data)
        else:
            res = get_data(key, value, res)
    
    return jsonify(res)


app.run(debug=False, use_reloader=False)
```
